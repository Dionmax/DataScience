### Naive Bayes

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
credito.drop(columns=["IDCREDITO"], inplace=True)
```

```python
credito.shape
```

```python
previsores = credito.iloc[:, 0:18].values
classe = credito.iloc[:, 18].values
```

```python
previsores
```

```python
X_Treino, X_Teste, Y_Treino, Y_Teste = train_test_split(
    previsores, classe, test_size=0.3, random_state=0
)
```

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_Treino, Y_Treino)
```

```python
previsoes = naive_bayes.predict(X_Teste)
```

```python
confusao = confusion_matrix(Y_Teste, previsoes)
confusao
```

```python
taxa_acerto = accuracy_score(Y_Teste, previsoes)
taxa_acerto
```

```python
print("Accuracy: {:.2f}%".format(taxa_acerto * 100))
```

### Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
```

```python
previsores = credito.iloc[:, 0:18].values
classe = credito.iloc[:, 18].values
```

```python
X_Treino, X_Teste, Y_Treino, Y_Teste = train_test_split(
    previsores, classe, test_size=0.2, random_state=0
)
```

```python
# Normalizing the data
sc = StandardScaler()

X_Treino = sc.fit_transform(X_Treino)
X_Teste = sc.transform(X_Teste)
```

```python
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units=9, activation="relu", kernel_initializer="uniform", input_dim=18))

# Adding the second hidden layer
classifier.add(Dense(units=9, activation="relu", kernel_initializer="uniform"))

# Adding the output layer
classifier.add(Dense(units=1, activation="sigmoid"))

# Compiling the ANN
classifier.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Training the ANN
classifier.fit(X_Treino, Y_Treino, batch_size=10, epochs=100)
```

```python
Y_prediction = classifier.predict(X_Teste)
Y_prediction = Y_prediction > 0.5
```

```python
confusao = confusion_matrix(Y_Teste, Y_prediction)

taxa_acerto = accuracy_score(Y_Teste, Y_prediction)

print("Accuracy: {:.2f}%".format(taxa_acerto * 100))
```
