```python
from skmultilearn.adapt import MLARAM
from skmultilearn.problem_transform import BinaryRelevance, ClassifierChain, LabelPowerset
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import hamming_loss
import pandas as pd
```

```python
df = pd.read_csv('data/musica.csv')

print(df.shape)
df.head()
```

```python
classe = df.iloc[:, 0:6].values
previsores = df.iloc[:, 7:78].values

classe
```

```python
X_train, X_test, y_train, y_test = train_test_split(previsores, classe, test_size=0.3, random_state=0)
```

```python
ann = MLARAM()

ann.fit(X_train, y_train)
```

```python
previsto = ann.predict(X_test)

# Hamming Loss
1 - hamming_loss(y_test, previsto)
```

```python
binary = BinaryRelevance(SVC())

binary.fit(X_train, y_train)

previsto = binary.predict(X_test)

# Hamming Loss
1 - hamming_loss(y_test, previsto)
```

```python
chain = ClassifierChain(SVC())

chain.fit(X_train, y_train)

previsto = chain.predict(X_test)

# Hamming Loss
1 - hamming_loss(y_test, previsto)
```

```python
label = LabelPowerset(SVC())

label.fit(X_train, y_train)

previsto = label.predict(X_test)

# Hamming Loss
1 - hamming_loss(y_test, previsto)
```
