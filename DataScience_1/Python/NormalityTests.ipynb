```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
data = norm.rvs(size=1000)

data
```

```python
plt.hist(data, bins=20)
plt.title('Histograma')
plt.show()
```

```python
fig, ax = plt.subplots()
stats.probplot(data, fit=True, plot=ax)
plt.show()
```

```python
# Teste de normalidade
# H0: A amostra vem de uma distribuição normal
# H1: A amostra não vem de uma distribuição normal
# Se p-value < 0.05, rejeita-se H0
# Se p-value > 0.05, não rejeita-se H0
# Se p-value < 0.05, a amostra não vem de uma distribuição normal
# Se p-value > 0.05, a amostra vem de uma distribuição normal

stats.shapiro(data)
```

```python
data2 = skewnorm.rvs(a=4, size=1000)

data2
```

```python
plt.hist(data2, bins=20)
plt.title('Histograma')
plt.show()
```

```python
fig, ax = plt.subplots()
stats.probplot(data2, fit=True, plot=ax)
plt.show()
```
