```r
model <- lm(speed ~ dist, data = cars)

model
```

```r
plot(speed ~ dist, data = cars, main = "Speed vs Distance", xlab = "Distance", ylab = "Speed")
abline(model, col = "red")
```

```r
model$coefficients
```

```r
model$coefficients[1] + model$coefficients[2] * 22
```
