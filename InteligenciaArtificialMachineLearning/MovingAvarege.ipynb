```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(0, 1, 100) + np.linspace(0, 1, 100)

data[25] = 7
data[50] = 6
```

```python
plt.plot(data)
plt.title("Line plot")
plt.show()
```

```python
s = pd.Series(data)

s_ma = s.rolling(window=10).mean()
```

```python
plt.plot(s_ma)
plt.title("Moving average")
plt.show()
```

```python
anomalies = s[np.abs(s - s.mean()) > 3 * s.std()]

anomalies
```

```python
plt.plot(s, label="Data")
plt.plot(s_ma, label="Moving average")
plt.plot(anomalies, "ro", markersize=8, label="Anomalies")
plt.legend()
plt.title("Anomalies detection")
plt.show()
```
