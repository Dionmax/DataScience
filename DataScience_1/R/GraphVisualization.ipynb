```r
library(igraph)
```

```r
grafo <- graph(edges = c("Fernando", "Pedro", "José", "Antônio", "Fernando", "José", "Fernando", "Antônio"))
V(grafo)$Peso <- c(40, 30, 30, 25)
E(grafo)$TipoAmizade <- c("Amigo", "Inimigo", "Inimigo", "Amigo")
E(grafo)$weight <- c(1, 2, 1, 3)
V(grafo)$type <- "Humanos"
```

```r
plot(grafo, vertex.size = V(grafo)$Peso)
```

```r
plot(grafo, vertex.size = V(grafo)$Peso, edge.width = edge_attr(grafo)$weight)
```

```r
vertex_attr(grafo)$Cor <- c("blue", "red", "green", "yellow")

plot(grafo, vertex.size = V(grafo)$Peso, edge.width = edge_attr(grafo)$weight, vertex.color = vertex_attr(grafo)$Cor)
```

```r
plot(grafo, vertex.size = V(grafo)$Peso, edge.width = edge_attr(grafo)$weight, vertex.color = vertex_attr(grafo)$Cor, edge.curved = 0.4, frame = TRUE)
```

```r
plot(grafo, vertex.size = V(grafo)$Peso, edge.width = edge_attr(grafo)$weight, vertex.color = vertex_attr(grafo)$Cor, edge.curved = 0.4, frame = TRUE, main = "Meu Grafo", vertex.shape = "square")
```
