```r
library(ggplot2)
library(cluster)
library(factoextra)
```

```r
cluster <- pam(iris[, 1:4], k = 3)
cluster
```

```r
plot(cluster)
```

```r
table(iris$Species, cluster$clustering)
```

```r
g <- fviz_cluster(list(data = iris[, 1:4], cluster = cluster$clustering), ellipse.type = "norm", ggtheme = theme_bw(), main = "PAM Clustering")
plot(g)
```

```r
g <- fviz_cluster(cluster, data = iris[, 1:4], geom = "point", ellipse.type = "convex", ellipse.level = 0.95, main = "PAM Clustering Iris Data Set")
plot(g)
```
