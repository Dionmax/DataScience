```r
suavizacao_exponencial <- ets(AirPassengers)
suavizacao_exponencial
```

```r
previsao <- forecast(suavizacao_exponencial, h = 12)
plot(previsao)
```

```r
# ARIMA(2,1,1)(0,1,0)[12]
# ARIMA(p,d,q)(P,D,Q)[s]
# p: ordem do modelo autorregressivo
# d: ordem de diferenciação
# q: ordem do modelo de médias móveis
# P: ordem do modelo autorregressivo sazonal
# D: ordem de diferenciação sazonal
# Q: ordem do modelo de médias móveis sazonal
# s: frequência sazonal

arima_model <- auto.arima(AirPassengers)
arima_model
```

```r
previsao_arima <- forecast(arima_model, h = 12)

plot(previsao_arima)
```
