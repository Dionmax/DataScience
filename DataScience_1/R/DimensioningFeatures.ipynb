```r
iris_padr <- scale(iris[, 1:4])
head(iris_padr, 10)
```

```r
normaliza <- function(x) {
    return((x - min(x)) / (max(x) - min(x)))
}
```

```r
iris_norm <- normaliza(iris[, 1:4])
head(iris_norm, 10)
```

```r
boxplot(iris[, 1:4], main = "Boxplot das variáveis originais")
boxplot(iris_padr, main = "Boxplot das variáveis padronizadas")
boxplot(iris_norm, main = "Boxplot das variáveis normalizadas")
```
