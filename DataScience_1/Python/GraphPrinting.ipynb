```python
from igraph import Graph
from igraph import plot
```

```python
grafo = Graph(edges=[(0, 1), (1, 2), (2, 3), (3, 0), (0, 2), (1, 3)], directed=True)

grafo.vs["label"] = ["Fernando", "Pedro", "Jose", "Antonio"]

grafo.vs["peso"] = [40, 30, 30, 25]

grafo.es["TipoAmizade"] = ["Amigo", "Inimigo", "Inimigo", "Amigo", "Amigo", "Amigo"]

grafo.es["weight"] = [1, 2, 1, 3, 5, 4]

print(grafo)

plot(grafo, margin=100)
```

```python
for v in grafo.vs:
    print(v)
```

```python
for e in grafo.es:
    print(e)
```

```python
grafo.vs["cor"] = ["blue", "red", "yellow", "green"]

plot(grafo, bbox=(300, 300), vertex_size=grafo.vs["peso"], vertex_color=grafo.vs["cor"])
```

```python
plot(
    grafo,
    bbox=(300, 300),
    vertex_size=grafo.vs["peso"],
    vertex_color=grafo.vs["cor"],
    edge_width=grafo.es["weight"],
    edge_curved=True,
)
```

```python
plot(
    grafo,
    bbox=(300, 300),
    vertex_size=grafo.vs["peso"],
    vertex_color=grafo.vs["cor"],
    edge_width=grafo.es["weight"],
    edge_curved=True,
    vertex_shape="square",
)
```
