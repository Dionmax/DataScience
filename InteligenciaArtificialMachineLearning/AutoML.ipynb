```python
import h2o
from h2o.automl import H2OAutoML
from h2o.estimators import H2ORandomForestEstimator
import pandas as pd
```

```python
h2o.init()
```

```python
data = h2o.import_file('data/insurance.csv')

data.head()
```

```python
data.drop('C1')
```

```python
train, test = data.split_frame(ratios=[0.7], seed=0)
```

```python
model_automl = H2OAutoML(max_runtime_secs=180, sort_metric='AUTO')

model_automl.train(y='Accident', training_frame=train)
```

```python
ranking = model_automl.leaderboard

ranking = ranking.as_data_frame()

ranking
```

```python
prediction = model_automl.leader.predict(test)

prediction
```
