```r
# install.packages("arules")
# install.packages("arulesViz")

library(arules)
library(arulesViz)
```

```r
transacoes <- read.transactions(file = "../data/transacoes.txt", format = "basket", sep = ",")
transacoes
```

```r
inspect(transacoes)
```

```r
image(transacoes, main = "Transações")
```

```r
regras <- apriori(transacoes, parameter = list(support = 0.5, confidence = 0.5, minlen = 2))
regras
```

```r
summary(regras)
```

```r
inspect(regras)
```

```r
plot(regras, method = "graph", control = list(type = "items"))
```

```r
plot(regras, method = "matrix", control = list(type = "items"))
```

```r
plot(regras, method = "Matrix3D")
```
