```python
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
```

```python
iris = datasets.load_iris()
```

```python
# Fuzzy c-means
r = skfuzzy.cmeans(data=iris.data.T, c=3, m=2, error=0.005, maxiter=1000, init=None)
```

```python
previsoes_porcentagem = r[1]
previsoes_porcentagem[0][0]
```

```python
for x in range(150):
    print(iris.target[x])
    print(previsoes_porcentagem[0][x])
    print(previsoes_porcentagem[1][x])
    print(previsoes_porcentagem[2][x])
    print('---')
```

```python
previsoes = previsoes_porcentagem.argmax(axis=0)
resultados = confusion_matrix(iris.target, previsoes)
print(resultados)
```

```python
taxa_acerto = np.sum(np.diagonal(resultados)) / np.sum(resultados)
print(taxa_acerto)
```
