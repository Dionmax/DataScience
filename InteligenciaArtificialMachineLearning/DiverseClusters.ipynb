```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = datasets.load_iris()

iris
```

```python
def plot_clusters(data, labels, title):
    colors = ["red", "green", "blue", "black"]
    plt.figure(figsize=(8, 4))

    for i, c, l in zip(
        range(-1, 3), colors, ["Noise", "Setosa", "Versicolor", "Virginica"]
    ):
        if i == -1:
            plt.scatter(
                data[labels == i, 0],
                data[labels == i, 3],
                c=colors[i],
                label=l,
                alpha=0.5,
                s=50,
                marker="x",
            )
        else:
            plt.scatter(
                data[labels == i, 0],
                data[labels == i, 3],
                c=colors[i],
                label=l,
                alpha=0.5,
                s=50,
            )
    
    plt.title(title)
    plt.legend()
    plt.xlabel('Sepal Length')
    plt.ylabel('Petal Width')
    plt.show()
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')

kmeans.fit(iris.data)

print(kmeans.labels_)
```

```python
plot_clusters(iris.data, kmeans.labels_, 'KMeans')
```

```python
iris.target
```

```python
print(confusion_matrix(iris.target, kmeans.labels_))
```

```python
db_scan = DBSCAN(eps=0.5, min_samples=3)

db_scan_labels = db_scan.fit_predict(iris.data)

print(db_scan_labels)
```

```python
plot_clusters(iris.data, db_scan_labels, 'DBSCAN')
```

```python
agglo = AgglomerativeClustering(n_clusters=3)

agglo_labels = agglo.fit_predict(iris.data)

print(agglo_labels)
```

```python
plot_clusters(iris.data, agglo_labels, 'Agglomerative')
```

```python
print(confusion_matrix(iris.target, agglo_labels))
```

```python
plt.figure(figsize=(12, 6))
plt.title('Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')

linkage_matrix = linkage(iris.data, method='ward')
dendrogram(linkage_matrix, truncate_mode='lastp', p=15)

# Define the threshold
plt.axhline(y=7, c='gray', lw=1, linestyle='dashed')
```
