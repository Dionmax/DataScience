```python
import matplotlib.pyplot as plt
import nltk

nltk.download("stopwords")

from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords

from matplotlib.colors import ListedColormap

# conda install -c conda-forge wordcloud
from wordcloud import WordCloud

import string
```

```python
corpus = PlaintextCorpusReader("../data/NLP_FILES", ".*(.txt)", encoding="ISO-8859-1")

arquivos = corpus.fileids()

arquivos[0:10]
```

```python
texto = corpus.raw("1.txt")

texto
```

```python
todo_texto = corpus.raw()
```

```python
palavras = corpus.words()

len(palavras)
```

```python
stops = stopwords.words("english")

# stops
```

```python
mapa_core = ListedColormap(["orange", "green", "red", "magenta"])

nuvem = WordCloud(
    background_color="white", colormap=mapa_core, stopwords=stops, max_words=100
)

nuvem.generate(todo_texto)

plt.imshow(nuvem)
```

```python
palavras_sem_stop = [p for p in palavras if p not in stops]
len(palavras_sem_stop)
```

```python
palavras_sem_ponct = [p for p in palavras_sem_stop if p not in string.punctuation]
len(palavras_sem_ponct)
```

```python
frequencia = nltk.FreqDist(palavras_sem_ponct)
frequencia
```

```python
mais_comuns = frequencia.most_common(100)
mais_comuns[:10]
```
