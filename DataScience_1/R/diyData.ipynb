```r
# install.packages("RColorBrewer")
library("RColorBrewer")
library(lattice)

dataset <- read.csv(
    "../data/dados.csv",
    sep = ";",
    na.strings = "",
    stringsAsFactors = TRUE
)
head(dataset, 10)
```

```r
dataset[!complete.cases(dataset), ]
```

```r
summary(dataset)
```

```r
hist(dataset$VALOREMPENHO, main = "Municipio", xlab = "Valor Empenho")
```

```r
boxplot(dataset$VALOREMPENHO)
```

```r
plot(dataset$VALOREMPENHO, dataset$VALORLIQUIDADO, main = "Valor Empenho x Valor Liquidado", xlab = "Valor Empenho", ylab = "Valor Liquidado")
```

```r
hist(dataset$PIB)
```

```r
plot(dataset$PIB, dataset$VALOREMPENHO, main = "PIB x Valor Empenho", xlab = "PIB", ylab = "Valor Empenho")
```

```r
PIBDensity <- density(dataset$PIB)
plot(PIBDensity, main = "Densidade do PIB", col="blue")
```

```r
ValorDensity <- density(dataset$VALOREMPENHO)
plot(ValorDensity, main = "Densidade do Valor Empenho", col = "red")
```

```r
plot(dataset$PIB, dataset$VALOREMPENHO, main = "PIB x Valor Empenho", xlab = "PIB", ylab = "Valor Empenho", pch = 20)
```

```r
summary(dataset$VALOREMPENHO)
summary(dataset$PIB)
```

```r
boxplot(dataset$VALOREMPENHO, main="Valores de Empenho", col = brewer.pal(n = 3, name = "Paired"), outline = F )
```

```r
boxplot(dataset$PIB ,main="Valores de PIB", col = brewer.pal(n = 3, name = "Pastel1"), outline = F )
```

```r
hist(dataset$VALOREMPENHO,main="Valores de Empenho", col = brewer.pal(n = 3, name = "Paired"))
```

```r
hist(dataset$PIB,main="Valores de PIB", col = brewer.pal(n = 3, name = "Pastel1"), outline = F )
```

```r
Mempenho <- head(dataset[order(-dataset$VALOREMPENHO), ], 10)
barplot(Mempenho$VALOREMPENHO, col = brewer.pal(n = 10, name = "RdBu"), las = 2, main = "Maiores Empenhos")
legend("topright", legend = Mempenho$MUNICIPIO, col = brewer.pal(n = 10, name = "RdBu"), lty = 1:2, cex = 0.8, ncol = 2, lwd = 4)
box()
```

```r
Mpibs <- head(dataset[order(-dataset$PIB), ], 10)
barplot(Mpibs$PIB, col = brewer.pal(n = 10, name = "Spectral"), las = 2, main = "Maiores PIBS")
legend("topright", legend = Mempenho$MUNICIPIO, col = brewer.pal(n = 10, name = "Spectral"), lty = 1:2, cex = 0.8, ncol = 2, lwd = 4)
box()
```

```r
dataset$PROPORCAO <- dataset$VALOREMPENHO / dataset$PIB
Mprop <- head(dataset[order(-dataset$PROPORCAO), ], 10)
barplot(Mprop$PROPORCAO, col = brewer.pal(n = 10, name = "Set2"), las = 2, main = "Maiores Gastos em Proporção ao PIB")
legend("topright", legend = Mprop$MUNICIPIO, col = brewer.pal(n = 10, name = "Set2"), lty = 1:2, cex = 0.8, ncol = 2, lwd = 4)
box()
```
