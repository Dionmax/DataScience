```r
# install.packages("factoextra")
library("factoextra")
library("ggplot2")
```

```r
head(iris, 10)
```

```r
cluster <- kmeans(iris[, -5], centers = 3)
cluster
```

```r
table(iris$Species, cluster$cluster)
```

```r
plot(iris$Sepal.Length, iris$Sepal.Width, col = cluster$cluster, pch = 20, main = "K-means with 3 clusters", xlab = "Sepal Length", ylab = "Sepal Width")
plot(iris[, 1:4], col = cluster$cluster, main = "K-means with 3 clusters")
```

```r
g2 <- fviz_cluster(cluster, data = iris[, -5], ellipse.type = "norm", ggtheme = theme_bw(), main = "K-means with 3 clusters")
plot(g2)
```

```r
g2 <- fviz_cluster(cluster, data = iris[, -5], ellipse.type = "convex", ggtheme = theme_bw(), main = "K-means with 3 clusters")
plot(g2)
```
