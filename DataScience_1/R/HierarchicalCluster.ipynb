```r
library(ggplot2)
library(factoextra)
```

```r
hc <- hcut(iris[, -5], k = 3, hc_method = "complete", stand = TRUE)
hc
```

```r
plot(hc, cex = 0.6, hang = -1, main = "Dendrogram of Iris Data")
```

```r
g <- fviz_dend(hc, show_labels = FALSE, rect = TRUE, ggtheme = theme_bw(), main = "Dendrogram of Iris Data")
plot(g)
```

```r
g1 <- fviz_cluster(hc, ellipse.type = "norm", ggtheme = theme_bw(), main = "Cluster of Iris Data")
plot(g1)
```

```r
g2 <- fviz_cluster(hc, data = iris[, -5], geom = "point", ellipse.type = "convex", ellipse.level = 0.68, main = "Hierarchical Clustering of Iris Data")
plot(g2)
```
