```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
base = pd.read_csv("../data/dados.csv", sep=";")

base.head(10)
```

```python
base.describe()
```

```python
sns.histplot(base.PIB, kde=True, bins=10).set(title="PIB")
```

```python
sns.histplot(base.sort_values("PIB", ascending=False)["PIB"].head(10), kde=True)
```

```python
sns.histplot(
    base.sort_values("VALOREMPENHO", ascending=False)["VALOREMPENHO"].head(10), kde=True
)
```
