```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
base = pd.read_csv("../data/co2.csv")

base.head(10)
```

```python
base.info
```

```python
x = base.conc
y = base.uptake
```

```python
uniques = list(set(base.Treatment))

uniques
```

```python
for i in range(len(uniques)):
    indice = base.Treatment == uniques[i]
    plt.scatter(x[indice], y[indice], label=uniques[i])
plt.legend(loc="lower right")
```

```python
sns.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.show()
```

```python
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
plt.figure(figsize=(10, 5))  # Ajuste o tamanho se necessário
plt.subplot(1, 2, 1)
sns.scatterplot(x=q.conc, y=q.uptake)
plt.title('Quebec')
# sns.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.subplot(1, 2, 2)
sns.scatterplot(x=m.conc, y=m.uptake)
plt.title('Mississippi')
plt.tight_layout()
plt.show()
```

```python
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x=ch.conc, y=ch.uptake)
plt.title('Chilled')
plt.subplot(1, 2, 2)
sns.scatterplot(x=nc.conc, y=nc.uptake)
plt.title('Non chilled')
plt.tight_layout()
plt.show()
```
