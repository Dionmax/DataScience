```r
data_frame <- read.csv(
  "../data/Churn.csv",
  sep = ";",
  na.strings = "",
  stringsAsFactors = TRUE
)
```

```r
head(data_frame, 15)
```

```r
summary(data_frame)
```

```r
colnames(data_frame) <- c(
    "Id",
    "Score",
    "State",
    "Gender",
    "Age",
    "Patrimony",
    "Balance",
    "Products",
    "Card",
    "Active",
    "Salary",
    "Churn"
)

head(data_frame, 15)
```

```r
counts <- table(data_frame$State)
barplot(counts, main = "States", xlab = "State", col = "blue")
```

```r
counts <- table(data_frame$Gender)
barplot(counts, main = "Gender", xlab = "Gender", col = "blue")
```

```r
summary(data_frame$Score)
boxplot(data_frame$Score, main = "Score", col = "blue")
hist(data_frame$Score, main = "Score", col = "blue")
```

```r
# Age

summary(data_frame$Age)
boxplot(data_frame$Age, main = "Age", col = "blue")
hist(data_frame$Age, main = "Age", col = "blue")
```

```r
# Balance

summary(data_frame$Balance)
boxplot(data_frame$Balance, main = "Balance", col = "blue")
hist(data_frame$Balance, main = "Balance", col = "blue")
```

```r
# Salary

summary(data_frame$Salary)
```

```r
# Data with null values

data_frame[!complete.cases(data_frame), ]
```

```r
# Treating null values in the Salary column

median(data_frame$Salary, na.rm = TRUE)

data_frame$Salary[is.na(data_frame$Salary)] <- median(data_frame$Salary, na.rm = TRUE)

data_frame[!complete.cases(data_frame$Salary),]
```

```r
# Treating null values in the Gender column

unique(data_frame$Gender)
summary(data_frame$Gender)

# Transforming the F in Feminino and M in Masculino
data_frame[is.na(data_frame$Gender) | data_frame$Gender == "M", ]$Gender <- "Masculino"
data_frame[data_frame$Gender == "F" | data_frame$Gender == "Fem", ]$Gender <- "Feminino"

summary(data_frame$Gender)

data_frame$Gender <- factor(data_frame$Gender)

summary(data_frame$Gender)
```

```r
# Treating outliers in the Age column
summary(data_frame$Age)

# Outlies
data_frame[data_frame$Age < 0 | data_frame$Age > 110, ]$Age

data_frame[is.na(data_frame$Age), ]

median(data_frame$Age)

data_frame[data_frame$Age < 0 | data_frame$Age > 110, ]$Age = median(data_frame$Age)

data_frame[data_frame$Age < 0 | data_frame$Age > 110, ]$Age

summary(data_frame$Age)
```

```r
# Duplicate data
x <- data_frame[duplicated(data_frame$Id), ]
x

# Removing duplicate data
data_frame <- data_frame[!duplicated(data_frame$Id), ]

x <- data_frame[duplicated(data_frame$Id), ]
x
```

```r
# States out of the pattern
unique(data_frame$State)
summary(data_frame$State)

# Fill with moda
data_frame[!data_frame$State %in% c("RS", "PR", "SC"), ]$State <- "RS"
summary(data_frame$State)

# Removing factors that are not in the pattern
data_frame$State <- factor(data_frame$State)
summary(data_frame$State)
```

```r
# Treatin outliers in the Salary column
summary(data_frame$Salary)

salary_sd <- sd(data_frame$Salary)
salary_sd

data_frame[data_frame$Salary >= 2 * salary_sd, ]$Salary

boxplot(data_frame$Salary)

# Outliers
# x = boxplot(data_frame$Salary)$out
# x

# Change the outliers to the median
salary_median <- median(data_frame$Salary)

data_frame[data_frame$Salary >= 2 * salary_sd, ]$Salary <- salary_median

data_frame[data_frame$Salary >= 2 * salary_sd, ]$Salary

boxplot(data_frame$Salary)
```
