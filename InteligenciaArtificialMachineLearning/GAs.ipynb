```python
# pip install pygad
import pygad

import numpy as np
```

```python
# Equantion: 2x + 5 = 20
def equantion_func(ga_instance, solution, solution_idx):
    x = solution[0]
    equantion_output = 2 * x + 5
    target_output = 20
    equantion_fitness = 1.0 / (np.abs(equantion_output - target_output) + 1e-8)
    return equantion_fitness


ga_instance = pygad.GA(
    num_generations=100,
    num_parents_mating=10,
    fitness_func=equantion_func,
    sol_per_pop=20,
    num_genes=1,
    init_range_low=-10,
    init_range_high=10,
)

ga_instance.run()
ga_instance.plot_fitness()

solution, solution_fitness, solution_idx = ga_instance.best_solution()

print("Parameters of the best solution :", round(solution[0], 4))
print("Fitness value of the best solution :", solution_fitness)
```
