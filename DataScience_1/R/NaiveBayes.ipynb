```r
library("e1071")
```

```r
credito <- read.csv(
    "../data/Credit.csv",
    sep = ",",
    na.strings = "",
    stringsAsFactors = TRUE
)
dim(credito)
head(credito, 10)
```

```r
# Transformando a variável target em fator
credito$class <- as.factor(credito$class)
```

```r
amostra <- sample(2, nrow(credito), replace = TRUE, prob = c(0.7, 0.3))
credito_treino <- credito[amostra == 1, ]
credito_teste <- credito[amostra == 2, ]
```

```r
dim(credito_treino)
dim(credito_teste)
```

```r
modelo <- naiveBayes(class ~ ., data = credito_treino)
modelo
```

```r
predicao <- predict(modelo, credito_teste)
predicao
```

```r
confusao <- table(credito_teste$class, predicao)
confusao
```

```r
taxa_acerto <- (confusao[1] + confusao[4]) / sum(confusao)
taxa_erro <- (confusao[2] + confusao[3]) / sum(confusao)
```
