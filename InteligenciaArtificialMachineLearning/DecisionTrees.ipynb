```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
)

import graphviz
```

```python
base = pd.read_csv("data/insurance.csv")

base.head()
```

```python
base.drop(columns=["Unnamed: 0"], inplace=True)

base.head()
```

```python
base.dropna(inplace=True)
```

```python
y = base.iloc[:, 7].values

X = pd.concat([base.iloc[:, 0:6], base.iloc[:, 8:]], axis=1)

X = X.values
```

```python
labelEncoder = LabelEncoder()
```

```python
for i in range(0, X.shape[1]):
    if X[:, i].dtype == "object":
        X[:, i] = labelEncoder.fit_transform(X[:, i])
```

```python
# X independent variables
# y dependent variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
modelo = DecisionTreeClassifier(random_state=1) # 79%
modelo = DecisionTreeClassifier(random_state=1, max_depth=8) # 84%
modelo = DecisionTreeClassifier(random_state=1, max_depth=8, max_leaf_nodes=8) # 86%

modelo.fit(X_train, y_train)
```

```python
previsao = modelo.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, previsao))
print("Precision: ", precision_score(y_test, previsao, average="micro"))
print("Recall: ", recall_score(y_test, previsao, average="micro"))
print("F1: ", f1_score(y_test, previsao, average="micro"))
```

```python
print(classification_report(y_test, previsao))
```
