```r
data_frame <- read.csv(
    "../data/Anova.csv",
    sep = ";",
    na.strings = "",
    stringsAsFactors = TRUE
)

head(data_frame, 5)
```

```r
boxplot(data_frame$Horas ~ data_frame$Remedio)
```

```r
an <- aov(data_frame$Horas ~ data_frame$Remedio)
summary(an)
```

```r
an = aov(data_frame$Horas ~ data_frame$Remedio * data_frame$Sexo)

summary(an)
```

```r
tukey <- TukeyHSD(an)

tukey
```

```r
plot(tukey)
```
