```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
base = pd.read_csv("../data/trees.csv")

base.shape
```

```python
base.head()
```

```python
h = np.histogram(base.iloc[:, 1], bins=6)
h
```

```python
plt.hist(base.iloc[:, 1], bins=6)
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel("Altura")
```

```python
sns.histplot(base.iloc[:, 1], kde=False, bins=6, color="blue").set_title("Árvores")
```

```python
sns.kdeplot(base.iloc[:, 1], color="blue").set_title("Árvores")
```

```python
sns.histplot(base.iloc[:, 1], kde=True, bins=6, color="blue").set_title("Árvores")
```

```python
plt.scatter(x=base.Girth, y=base.Volume, color="blue", marker="*", facecolors="none")
plt.title("Árvores")
plt.xlabel("Circunferência")
plt.ylabel("Volume")
```

```python
plt.plot(base.Girth, base.Volume)
plt.title("Árvores")
plt.xlabel("Circunferência")
plt.ylabel("Volume")
```
