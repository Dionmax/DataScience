```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
# conda install conda-forge::mlxtend
# conda install conda-forge/label/cf201901::mlxtend
# conda install conda-forge/label/cf202003::mlxtend
# conda install conda-forge/label/gcc7::mlxtend
# conda install -c conda-forge mlxtend 

from mlxtend.plotting import plot_decision_regions
```

#### Linear Model

```python
# Change max_iter to 10000 to avoid warnings
modelo = linear_model.LogisticRegression(solver='lbfgs', max_iter=10000)

carros = pd.read_csv('../data/carros.csv', sep=';')

carros.head()
```

```python
carros = carros.to_numpy()
```

```python
y = carros[:, 9]
X = carros[:, 0:2]
```

```python
modelo.fit(X, y)
```

```python
plt.figure(figsize=(12,8))

fig = plot_decision_regions(X=X, y=y, clf=modelo)
plt.title("Limites de Decisão")
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left", ncol=2)
plt.xlabel("Consumo")
plt.ylabel("Cilindros")
plt.show()
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

modelo = RandomForestClassifier(random_state=0)
```

```python
modelo.fit(X, y)
```

```python
plt.figure(figsize=(12,8))

fig = plot_decision_regions(X=X, y=y, clf=modelo)
plt.title("Limites de Decisão")
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left", ncol=2)
plt.xlabel("Consumo")
plt.ylabel("Cilindros")
plt.show()
```
