```r
cor(mtcars[1:4])
corrgram(mtcars[1:4], order = TRUE, lower.panel = panel.shade, upper.panel = panel.pie, text.panel = panel.txt, main = "Correlogram of mtcars")
```

```r
corrgram(mtcars, order = TRUE, lower.panel = panel.shade, upper.panel = panel.pie, text.panel = panel.txt, main = "Correlogram of mtcars")
```

```r
plot(mtcars$mpg ~ mtcars$disp, ylab = "Miles per gallon", xlab = "Displacement", main = "Scatterplot of mtcars")
```

```r
modelo <- lm(mpg ~ disp, data = mtcars)
modelo
```

```r
summary(modelo)$r.squared
```

```r
summary(modelo)$adj.r.squared
```

```r
plot(mtcars$mpg ~ mtcars$disp, ylab = "Miles per gallon", xlab = "Displacement", main = "Scatterplot of mtcars")
abline(modelo, col = "red")
```

```r
predict(modelo, data.frame(disp = 200))
```

```r
modelo <- lm(mpg ~ disp + hp + cyl, data = mtcars)
modelo
```

```r
summary(modelo)$r.squared
```

```r
summary(modelo)$adj.r.squared
```

```r
predict(modelo, data.frame(disp = 200, hp = 100, cyl = 4))
```
